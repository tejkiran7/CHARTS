# Distribution plots

1. distplot<br>
2. jointplot<br>
3. pairplot<br>

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

## Correlation with heatmap

```python
df.corr() #values range between -1 to +1
```

```python
sns.heatmap(df.corr())
```

```python
df.dtypes
```

<b>Univariate Analysis</b>

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

<b>Pair plot</b>

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

<b>Dist Plot</b>

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots

1. boxplot<br>
2. violinplot<br>
3. countplot<br>
4. barplot

```python
# countplot

sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
##Bar plot

sns.barplot(x='total_bill',y='sex',data=df)
```

## Box Plot 

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(y='day',x='total_bill',hue='smoker',data=df)
```

## Violin Plot

```python
sns.violinplot(y='day',x='total_bill',data=df,palette='rainbow')
```

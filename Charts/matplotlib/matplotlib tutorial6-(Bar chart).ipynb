```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","Car","other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,radius=1.5,autopct="%0.2f%%",shadow=True,explode=[0,0.2,0,0,0],startangle=180)
#plt.show()
plt.savefig("piechart.png",bbox_inches="tight",pad_inches=2,transparent=True)
```
